```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

df=pd.read_csv("/Users/preetamkamate/teams.csv")
df=df.fillna(0)
print(df.iloc[0:5,0:5])
```

```python
x=df[["prev_medals","athletes","age"]]
y=df[["medals"]]
x_train,x_test,y_train, y_test=train_test_split(x, y,test_size=0.2,random_state=42)
model=LinearRegression()
model.fit(x_train,y_train)
y_pred=model.predict(x_test)


country=input("enter the country name")
data= df[df["country"] == country][["prev_medals", "athletes", "age"]]
predicted= model.predict(data)
print("country:",country,"year",2025,":",int(predicted[0].item()),"medals")
```
